```python
state_text = [item.text for item in soup.findAll(class_ = 'bg-white b--brand-lighter-gray bt f8 v-top pa3 sticky left-0 z-1')]
state_text
```

```python
fundingmodel_tag = soup.findAll(class_ = 'b--brand-lighter-gray bt f8 v-top pa3')
fundingmodel_tag
#it's not just the funding model that uses this tag and class
```

```python
additional_rows = [item.text for item in soup.findAll(class_ = 'b--brand-lighter-gray bt f8 v-top pa3')]
additional_rows
#creates a similar output to the pd.read_html but it is organized by row rather than by column
```
